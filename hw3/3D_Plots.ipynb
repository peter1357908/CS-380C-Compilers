```python
from mpl_toolkits import mplot3d
```

```python
# %matplotlib inline
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

Calculate the **Jacobi** estimates

```python
NUM_ITERS = 10
iters = np.array([i for i in range(1, 1+NUM_ITERS)])
x_estimates = np.zeros(NUM_ITERS)
y_estimates = np.zeros(NUM_ITERS)

# we only need to calculate iteration 2+, since iteration 1 is given (zeros)
for i in range(2, 1+NUM_ITERS):
    x_estimates[i-1] = 4/3 - y_estimates[i-2]/3
    y_estimates[i-1] = 3/2 - x_estimates[i-2]/2

print(np.array2string(x_estimates, precision=3, separator=', '))
print(np.array2string(y_estimates, precision=3, separator=', '))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x_estimates, y_estimates, iters);

# the precise answer, stretched out to include the 0-level
x_precises = np.ones(NUM_ITERS)
y_precises = np.ones(NUM_ITERS)
z_precises = np.array([i for i in range(1, 1+NUM_ITERS)])
ax.plot3D(x_precises, y_precises, z_precises, c='red');
```

Calculate the **Gauss-Seidel** estimates

```python
x_estimates_GS = np.zeros(NUM_ITERS)
y_estimates_GS = np.zeros(NUM_ITERS)
# we only need to calculate iteration 2+, since iteration 1 is given (zeros)
for i in range(2, 1+NUM_ITERS):
    x_estimates_GS[i-1] = (4 - y_estimates_GS[i-2])/3
    y_estimates_GS[i-1] = (5 + y_estimates_GS[i-2])/6

print(np.array2string(x_estimates_GS, precision=3, separator=', '))
print(np.array2string(y_estimates_GS, precision=3, separator=', '))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x_estimates_GS, y_estimates_GS, iters);

# the precise answer
ax.plot3D(x_precises, y_precises, z_precises, c='red');
```
